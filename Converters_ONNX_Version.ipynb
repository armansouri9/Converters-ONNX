```python
import onnx
from onnx import version_converter
import torch
from onnx2torch import convert

# Load the ONNX model.
model = onnx.load('/content/dr_kiae1.onnx')
# Convert the model to the target version.
target_version = 14
converted_model = version_converter.convert_version(model, target_version)
# Convert to torch.
torch_model1 = convert(converted_model)
torch.save(torch_model1, '/content/modeldr_kiae1.pt')
torch_model1
```

```python
import onnx
from onnx import version_converter
import torch
from onnx2torch import convert

# Load the ONNX model.
model = onnx.load('/content/best.onnx')
# Convert the model to the target version.
target_version = 14
converted_model = version_converter.convert_version(model, target_version)
# Convert to torch.
torch_model = convert(converted_model)

torch.save(torch_model, '/content/model.pt')

torch_model
```

```python
random_input = torch.randn(1, 3, 640, 640, requires_grad=True)

torch.onnx.export(
  model=torch_model,
  args=random_input, 
  f="alireza.onnx", # where should it be saved

  input_names=['images'],
  output_names=['output0'],opset_version=13
)
```
